```python
import pandas as pd

score_file = "../fe2o3_bank_5_logit_results.csv"

df = pd.read_csv(score_file, index_col=0)

# Remove outliers
for i, idx in enumerate([20, 48, 119]):
    df = df.drop(df.columns[idx - i], axis=1)


temperature = [float(h.split("_")[2].split("-")[0]) for h in df.columns.tolist()]

plt.figure(figsize=(10, 5))

imax = 75

for i in range(5):
    plt.plot(
        temperature[:imax],
        df.iloc[i].to_numpy()[:imax],
        marker="o",
        linestyle="-",
        label=f"Row {i + 1}",
    )

# for i in range(len(df.columns.tolist())):
#    print(f'{i} {df.columns[i]}: {df.iloc[4, i]}')
```
